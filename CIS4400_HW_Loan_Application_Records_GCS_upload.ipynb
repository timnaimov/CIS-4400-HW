```python
import gdown
import zipfile
import pandas as pd
```

```python
import os.path
from os import path
```

```python
# Define the URL of the ZIP file
zip_url = "https://s3.amazonaws.com/cfpb-hmda-public/prod/dynamic-data/2022/2022_lar.zip"

# Define the path where you want to save the ZIP file
zip_file_path = "2022_lar.zip"

# Check if the ZIP file exists
if not path.exists(zip_file_path):
    # Download the ZIP file using gdown if it doesn't exist
    gdown.download(zip_url, zip_file_path, quiet=False)

# Extract the contents of the ZIP file if they haven't been extracted
with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
    # Extract the first file in the ZIP archive
    file_name = zip_ref.namelist()[0]
```

```python
# Define the number of rows to read as a sample
sample_size = 1000  # Adjust this number to the desired sample size

# Check if the sample DataFrame has been loaded
if 'df' not in locals():
    # Load a sample of the data into a DataFrame if it hasn't been loaded
    df = pd.read_fwf(file_name, nrows=sample_size)

    # Convert DataFrame to CSV format
    csv_file_path = "2022_lar_sample.csv"
    df.to_csv(csv_file_path, index=False)

    print("Conversion completed. CSV file saved as:", csv_file_path)
```

```python
# Check if the full DataFrame has been loaded
if 'df_full' not in locals():
    # Load the entire dataset into a DataFrame if it hasn't been loaded
    df_full = pd.read_fwf(file_name)

    # Convert DataFrame to CSV format
    csv_file_path_full = "2022_lar_full.csv"
    df_full.to_csv(csv_file_path_full, index=False)

    print("Full dataset conversion completed. CSV file saved as:", csv_file_path_full)
```
