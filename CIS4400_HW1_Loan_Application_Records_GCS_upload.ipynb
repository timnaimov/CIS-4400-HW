```python
# Import libraries
import pandas as pd
import numpy as np
import requests
import json
from io import StringIO
```

```python
# Download the file locally first
URL = "https://storage.cloud.google.com/2022_lar/2022_LAR_raw.txt"
response = requests.get(URL)

if response.status_code == 200:
    with open('2022_LAR_raw.txt', 'wb') as file:
        file.write(response.content)
    print("TXT file downloaded locally.")
else:
    print(f"Failed to download the file. Status code: {response.status_code}")
```

```python
# Load the CSV file into a DataFrame
df = pd.read_fwf('2022_LAR_raw.txt')
```

```python
from google.cloud import storage
```

```python
# Specify the name of your bucket
bucket_name = '2022_lar'

#Specify the path to your JSON configuration file
config_file_path = '/content/citric-sprite-418620-0ee5c498cc7a.json'

# Authenticate using a service account key
storage_client = storage.Client.from_service_account_json('/content/citric-sprite-418620-0ee5c498cc7a.json')

# Access the specified bucket
bucket = storage_client.bucket(bucket_name)

#specify name of the CSV file in the bucket
blob = bucket.blob (f'2022_LAR_raw.txt')

# Convert DataFrame to CSV
output = StringIO()
df.to_csv(output, index = False)
data = output.getvalue()
output.close()

try:
    # Upload CSV file to the bucket
    blob.upload_from_string(data, content_type='text/csv')
    print("File uploaded successfully!")
except Exception as e:
    print(f"Error uploading file: {e}")
```
