Data Source

```python
import gdown
import zipfile
import pandas as pd
```

```python
import os.path
from os import path
```

```python
# Define the URL of the ZIP file
zip_url = "https://s3.amazonaws.com/cfpb-hmda-public/prod/dynamic-data/2022/2022_lar.zip"

# Define the path where you want to save the ZIP file
zip_file_path = "2022_lar.zip"

# Check if the ZIP file exists
if not path.exists(zip_file_path):
    # Download the ZIP file using gdown if it doesn't exist
    gdown.download(zip_url, zip_file_path, quiet=False)

# Extract the contents of the ZIP file if they haven't been extracted
with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
    # Extract the first file in the ZIP archive
    file_name = zip_ref.namelist()[0]
    zip_ref.extract(file_name)
```

```python
# Define the chunk size for reading the CSV file
chunk_size = 100000  # Adjust as needed

# Initialize an empty list to store the sampled chunks
sampled_chunks = []

# Load the CSV file in chunks and sample from each chunk
for chunk in pd.read_csv(file_name, delimiter='|', chunksize=chunk_size, dtype=str, low_memory=False):
    sampled_chunk = chunk.sample(n=1000)  # Adjust the sample size as needed
    sampled_chunks.append(sampled_chunk)

# Concatenate the sampled chunks into a single DataFrame
df = pd.concat(sampled_chunks)

# Convert DataFrame to CSV format for further processing
csv_file_path_sample = "2022_lar_sample.csv"
df.to_csv(csv_file_path_sample, index=False)

print("Sample data conversion completed. CSV file saved as:", csv_file_path_sample)
```
